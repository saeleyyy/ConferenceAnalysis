```python
cursor = mydb.cursor()
cursor.execute("SHOW TABLES")
tables = cursor.fetchall()
tables
```

```python
emptyTables = []
tables_list = []
for table in tables:
    sql = "SELECT COUNT(*) FROM " + table[0]
    cursor.execute(sql)
    num_rows = cursor.fetchall()[0][0]
    tables_list.append([table[0], num_rows])
    if num_rows == 0:
        emptyTables.append(table[0])


tablesDF = pd.DataFrame(tables_list, columns=["Table", "Rows"])
print("*** Empty Tables ***")
print(tablesDF[tablesDF["Rows"] == 0])
print("*** Tables with Entries ***")
print(tablesDF[tablesDF["Rows"] != 0])
```
